```python
class List_Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class Linked_List:
    def __init__(self):
        self.head = None

    def insert(self, data):
        new_node = List_Node(data)
        if self.head is None:
            self.head = new_node
        else:
            node = self.head
            while node.next:
                node = node.next
            node.next = new_node

    def times_occurred(self, NUM):
        count = 0
        current = self.head
        while current:
            if current.data == NUM:
                count += 1
            current = current.next
        return count

    def display(self):
        current = self.head
        while current:
            print(current.data, end=" -> ")
            current = current.next
        print("None")
    def count_zero_nonzero(self):
        zero_count = 0
        nonzero_count = 0
        current = self.head
        while current:
            if current.data == 0:
                zero_count += 1
            else:
                nonzero_count += 1
            current = current.next
        return zero_count, nonzero_count
    
    def add_to_all(self, k):
        current = self.head
        while current:
            current.data += k
            current = current.next
    
# Example usage:
my_list = Linked_List()
my_list.insert(3)
my_list.insert(2)
my_list.insert(4)
my_list.insert(5)
my_list.insert(0)
my_list.insert(2)
my_list.insert(2)
my_list.insert(0)
my_list.insert(3)
my_list.insert(3)

target_item = 3
occurrences = my_list.times_occurred(target_item)
print(f"The item {target_item} occurs {occurrences} times in the linked list.")

(my_list.display())  # Print the linked list
print(my_list.times_occurred(3))

zero_count, nonzero_count = my_list.count_zero_nonzero()
print(f"Zero count: {zero_count}")
print(f"Nonzero count: {nonzero_count}")

k = 10
print(my_list.add_to_all(k))
print(my_list.display())
```
