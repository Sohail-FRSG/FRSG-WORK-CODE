```python
class Tree_Node:
    def __init__(self, value):
        self.left = None
        self.data = value
        self.right = None
class Tree:
    def create_node(self, data):
        return Tree_Node(data)
    def insert(self, node, data):
        if node is None:
            return self.create_node(data)
        if data < node.data:
            node.left = self.insert(node.left, data)
        else:
            node.right = self.insert(node.right, data)
        return node
    def display(self, node):
        if node:
            # Traverse the left subtree
            self.display(node.left)
            # Print the current node's data
            print(node.data, end=" ")
            # Traverse the right subtree
            self.display(node.right)

            
    def find_location(self, node, item):
        if node is None:
            return []  # item not found, return an empty path
        if item == node.data:
            return [node.data]  # item found, return a list with the item
        elif item < node.data:
            path = self.find_location(node.left, item)
            if path:
                return [node.data] + path
            else:
                return []
        else:
            path = self.find_location(node.right, item)
            if path:
                return [node.data] + path
            else:
                return []
        
    def count_nodes(self, node):
        if node is None:
            return 0
        return 1 + self.count_nodes(node.left) + self.count_nodes(node.right)

    def find_depth(self, node, target, depth=0):
        if node is None:
            return -1  # Target not found, return -1 to indicate it's not in the tree
        if target == node.data:
            return depth
        left_depth = self.find_depth(node.left, target, depth + 1)
        if left_depth != -1:
            return left_depth
        right_depth = self.find_depth(node.right, target, depth + 1)
        return right_depth
    
    def count_terminals(self, node):
        if node is None:
            return 0
        if node.left is None and node.right is None:
            return 1
        return self.count_terminals(node.left) + self.count_terminals(node.right)
    
    def delete_terminals(self, node):
        if node is None:
            return None
        if node.left is None and node.right is None:
            return None  
# Delete terminal nodes
        node.left = self.delete_terminals(node.left)
        node.right = self.delete_terminals(node.right)
        return node
    
    
    
data = [5, 2, 4, 10, 12, 7, 16, 3,19]            
obj = Tree()
root = obj.create_node(8)
root = None  # Initialize the root as None

# Insert data from the list into the tree
for value in data:
    root = obj.insert(root, value)
print(obj.display(root))
        
    
locaton_of_value = 3
path = obj.find_location(root, locaton_of_value)
if path:
    print(f"\nPath to target {locaton_of_value}: {path}")
else:
    print(f"\nTarget {locaton_of_value} not found in the tree.")
    
node_count = obj.count_nodes(root)
print(f"Number of nodes in the tree: {node_count}")

# Find the depth of a specific target node
target_node = 10
depth = obj.find_depth(root, target_node)
if depth != -1:
    print(f"Depth of target node {target_node}: {depth}")
else:
    print(f"Target node {target_node} not found in the tree.")
    
    
terminal_count = obj.count_terminals(root)
print(f"Number of terminal nodes in the tree: {terminal_count}")

# Delete all terminal nodes in the tree
root = obj.delete_terminals(root)

# Display the tree after deleting terminal nodes
print("\nTree after deleting terminal nodes:")
obj.display(root)
```
